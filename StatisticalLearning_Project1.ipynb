####4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points) Goal_diff_count = GoalsFor - GoalsAgainst

```
Laliga.dtypes
```

```
#Defining function

def Goal_diff_count(GoalsFor,GoalsAgainst):
    return (GoalsFor - GoalsAgainst)
```

```
row_col = Laliga.shape
max_row = row_col[0]
max_row
```

```
print(Laliga.GoalsFor.tail())
print(Laliga.GoalsAgainst.tail())
```

```
Goal_diff_count_List=[]
i=0
while (i < max_row):
    i = i+1
    GoalsFor = Laliga.loc[i,"GoalsFor"]
    GoalsAgainst = Laliga.loc[i,"GoalsAgainst"]
    x = Goal_diff_count(GoalsFor,GoalsAgainst)
    Goal_diff_count_List.append(x)
Laliga["Goal_diff_count_List"] = Goal_diff_count_List
```

```
Laliga[["Team","Goal_diff_count_List"]].head(10)
```

```
#Team with Maximum goal difference

Max_GD = Laliga["Goal_diff_count_List"].max()
print(Laliga[["Team","Goal_diff_count_List"]][Laliga.Goal_diff_count_List == Max_GD])
```

```
#Team with Minimum goal difference

Min_GD = Laliga["Goal_diff_count_List"].min()
print(Laliga[["Team","Goal_diff_count_List"]][Laliga.Goal_diff_count_List == Min_GD])
```

####5.Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

```
Winning_Percent=[]
i=0
while (i < max_row):
    i = i+1
    GamesWon = Laliga.loc[i,"GamesWon"]
    GamesPlayed = Laliga.loc[i,"GamesPlayed"]
    Percentage_of_Winning = (GamesWon / GamesPlayed)*100
    Percentage_of_Winning = Percentage_of_Winning.astype(int)
    if Percentage_of_Winning > 0:
        Percentage_of_Winning = (str(Percentage_of_Winning) + "%")
    else:    
        Percentage_of_Winning = "0%"
    Winning_Percent.append(Percentage_of_Winning)
Laliga["Winning Percent"] = Winning_Percent
```

###6.Print the top 5 teams which has the highest Winning percentage

Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
Eg: Best Position                Points

    1                  25000

    2                  7000      

```
Winning_Percentage=[]
i=0
while (i < max_row):
    i = i+1
    GamesWon = Laliga.loc[i,"GamesWon"]
    GamesPlayed = Laliga.loc[i,"GamesPlayed"]
    Winning_Percent = (GamesWon / GamesPlayed)*100
    Winning_Percent = Winning_Percent.astype(int)
    Winning_Percentage.append(Winning_Percent)
Laliga["Winning_Percentage"] = Winning_Percentage
Laliga_Winning_Percentage = Laliga.sort_values('Winning_Percentage', ascending=False)
print("Top 5 teams which has the highest Winning percentage are as below") 
Laliga_Winning_Percentage[['Team','Winning_Percentage']].head()
   
```
